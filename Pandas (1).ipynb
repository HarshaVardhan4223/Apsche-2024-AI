## Merge

```python
dd = pd.DataFrame({'Name':['Alina','Anu','Akhila','Dinesh','Siri'],'English':[78,81,92,93,94]})
dd
```

```python
dm = pd.DataFrame({'Name':['Alina','Anu','Akhila','Dinesh','Siri'],'Maths':[81,84,90,95,91]})
dm
```

```python
dh = pd.DataFrame({'Name':['Alina','Anu','Akhila','Dinesh','Siri'],'History':[89,88,90,92,91]})
dh
```

```python
m1 = pd.merge(dd,dm, on ='Name')
m1
```

```python
m2 = pd.merge(m1,dh, on ='Name')
```
